```python
from diffusers import StableDiffusionPipeline
import matplotlib.pyplot as plt
import torch
```

```python
model_id1='stable-diffusion-v1-5/stable-diffusion-v1-5'
model_id2='stability/stable-diffusion-x1-base-1.0'

pipe=StableDiffusionPipeline.from_pretrained(model_id1,torch_dtype=torch.float16, use_safetensors=True)
pipe.to('cuda')
```

```python
prompt='a boy in astronaut outfit landing on earth from its spaceship'
```

```python
image=pipe(prompt).images[0]
```

```python
image
```

```python
print('PROMPT:', prompt)
plt.imshow(image);
plt.axis('off');
```

```python
prompt2='A girl is carrying her white and black colour colured kitten and sitting on the chair'

image2=pipe(prompt2).images[0]
```

```python
print('PROMPT:', prompt2)
plt.imshow(image2);
plt.axis('off');
```
